```python
#Analyze the possible host vulnerabilities on 21, 22, 23 tcp port and interactions with mail servers.
import protocols.tcp.tcp as tcp
import protocols.fdns.type_txt.standard.dkim as dkim
import protocols.fdns.type_txt.standard.dmarc as dmarc
import protocols.fdns.type_txt.standard.sts as sts
import protocols.fdns.type_txt.standard.spf as spf
import protocols.fdns.type_a.a as a

"""
#Create necessary dataframe.
df_a = a_module.get_df_a('./data/project_sonar/2021-04-23-1619136719-fdns_a.json.gz')
df_ssh = tcp_module.get_df_tcp('./data/project_sonar/2021-04-29-1619726956-ssh_22.csv.gz')
df_telnet = tcp_module.get_df_tcp('./data/project_sonar/2021-04-30-1619776121-telnet_23.csv.gz')
df_ftp = tcp_module.get_df_tcp('./data/project_sonar/2021-04-07-1617768240-ftp_21.csv.gz')
df_dkim = dkim_module.get_df_dkim('./data/project_sonar2021-04-23-1619217645-fdns_txt.json.gz')
df_dmarc = dmarc_module.get_df_dmarc('./data/project_sonar/2021-04-23-1619217645-fdns_txt.json.gz')
df_sts = sts_module.get_df_sts('./data/project_sonar/2021-04-23-1619217645-fdns_txt.json.gz')
df_spf = spf_module.get_df_spf('./data/project_sonar/2021-04-23-1619217645-fdns_txt.json.gz')
"""

#Create necessary TEST dataframe.
df_a = a.get_df_a('./data/project_sonar/test_a.json')
df_ssh = tcp.get_df_tcp('./data/project_sonar/test_ssh.json')
df_telnet = tcp.get_df_tcp('./data/project_sonar/test_telnet.json')
df_ftp = tcp.get_df_tcp('./data/project_sonar/test_ftp.json')
df_dkim = dkim.get_df_dkim('./data/project_sonar/test_dkim.json')
df_dmarc = dmarc.get_df_dmarc('./data/project_sonar/test_dmarc.json')
df_sts = sts.get_df_sts('./data/project_sonar/test_sts.json')
df_spf = spf.get_df_spf('./data/project_sonar/test_spf.json')

#The hostname to search.
hostname = '0.188.12.93.rev.sfr.net'

#Find hostname in the A-dataframe.
df_a_filtered = a.fdns.get_df_hostname_specified(df_a, hostname)

#Find ip-address in all TCP-dataframes. 
ip_address = [df_a_filtered.iloc[0]['Ip']]
df_ssh_filtered = tcp.get_df_ip_selected(df_ssh, 'S_address', ip_address, 0)
df_telnet_filtered = tcp.get_df_ip_selected(df_telnet, 'S_address', ip_address, 0)
df_ftp_filtered = tcp.get_df_ip_selected(df_ftp, 'S_address', ip_address, 0)


#Find hostname in the TXT-dataframe.
df_dkim_filtered = dkim.txt.fdns.get_df_hostname_specified(df_dkim, hostname)
df_dmarc_filtered = dmarc.txt.fdns.get_df_hostname_specified(df_dmarc, hostname)
df_sts_filtered = sts.txt.fdns.get_df_hostname_specified(df_sts, hostname)
df_spf_filtered = a.fdns.handle.pd.DataFrame()
for index, item in enumerate(df_spf['IP4_ADDRESS']):
    if type(item) == list and ip_address[0] in item:
        df_spf_filtered = df_spf_filtered.append(df_spf.loc[index])
for index, item in enumerate(df_spf['IP6_ADDRESS']):
    if type(item) == list and ip_address[0] in item:
        df_spf_filtered = df_spf_filtered.append(df_spf.loc[index])
                     
#Output dataframe
print("SSH PORT OPEN:")
display(df_ssh_filtered)
print("\nTELNET PORT OPEN:")
display(df_telnet_filtered)
print("\nFTP PORT OPEN:")
display(df_ftp_filtered)
print("\nDKIM STANDARD:")
display(df_dkim_filtered)
print("\nDMARC STANDARD:")
display(df_dmarc_filtered)
print("\nSTS STANDARD:")
display(df_sts_filtered)
print("\nSPF STANDARD:")
display(df_spf_filtered)
```
